```python
import pandas as pd
```

```python
df = pd.read_json("TelecomX_Data.json")
```

```python
df.head()      # Muestra las primeras 5 filas
df.info()      # Muestra info de columnas, tipos de datos y nulos
```

```python

# Expandir cada columna anidada
customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])

# Combinar todo en un nuevo DataFrame
df_full = pd.concat(
    [df[['customerID', 'Churn']], customer_df, phone_df, internet_df, account_df],
    axis=1
)

# Verificamos cómo quedó
df_full.head()
```

```python
# Conteo de valores nulos por columna
df_full.isnull().sum()
```

```python
cat_cols = df_full.select_dtypes(include='object').columns
```

```python
for col in cat_cols:
    df_full[col] = df_full[col].str.strip().str.lower()
```

```python
# Lista de columnas categóricas para limpiar (strings)
cols_to_clean = [
    'Churn', 'gender', 'Partner', 'Dependents', 'PhoneService',
    'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
    'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaperlessBilling', 'PaymentMethod'
]

# Limpiar espacios y poner en minúsculas
for col in cols_to_clean:
    df_full[col] = df_full[col].str.strip().str.lower()

# Reemplazar valores largos o inconsistentes en PaymentMethod
df_full['PaymentMethod'] = df_full['PaymentMethod'].replace({
    'bank transfer (automatic)': 'bank_transfer',
    'credit card (automatic)': 'credit_card',
    'electronic check': 'e_check',
    'mailed check': 'mailed_check'
})

# Verificar cómo quedó
df_full['PaymentMethod'].unique()
```

```python
# Estadísticas básicas para las variables numéricas
df_full.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contar valores de churn
churn_counts = df_full['Churn'].value_counts(normalize=True)

# Plot
plt.figure(figsize=(6,4))
sns.barplot(x=churn_counts.index, y=churn_counts.values, palette='pastel')

plt.title('Proporción de Clientes por Estado de Churn')
plt.xlabel('Churn')
plt.ylabel('Proporción')
plt.ylim(0,1)

# Mostrar porcentaje encima de cada barra
for i, v in enumerate(churn_counts.values):
    plt.text(i, v + 0.02, f"{v:.2%}", ha='center', fontsize=12)

plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Variables categóricas para analizar
cat_vars = ['gender', 'Contract', 'PaymentMethod', 'InternetService', 'PhoneService', 'SeniorCitizen', 'PaperlessBilling']

plt.figure(figsize=(15, 10))

for i, var in enumerate(cat_vars, 1):
    plt.subplot(3, 3, i)

    # Calculamos la tasa de churn (proporción de 'yes') por categoría
    churn_rate = df_full.groupby(var)['Churn'].apply(lambda x: (x == 'yes').mean())

    # Barplot
    sns.barplot(x=churn_rate.index, y=churn_rate.values, palette='muted')
    plt.title(f'Tasa de Churn según {var}')
    plt.ylabel('Tasa de Churn')
    plt.ylim(0, 1)
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_vars = ['tenure', 'Charges.Monthly', 'Charges.Total']

plt.figure(figsize=(15, 4))

for i, var in enumerate(num_vars, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(x='Churn', y=var, data=df_full, palette='Set2')
    plt.title(f'Distribución de {var} según Churn')
    plt.xlabel('Churn')
    plt.ylabel(var)

plt.tight_layout()
plt.show()
```

```python
# Informe de Análisis de Evasión de Clientes (Churn) en TelecomX

## Introducción

El objetivo de este análisis es comprender y reducir la evasión de clientes (Churn) en TelecomX mediante el examen de datos de clientes. El Churn representa la pérdida de clientes y es un problema crítico para las empresas de telecomunicaciones, ya que adquirir nuevos clientes suele ser más costoso que retener los existentes.

A través de este análisis, buscamos:
- Identificar patrones y características comunes entre los clientes que abandonan el servicio
- Determinar los factores más influyentes en la decisión de cancelación
- Proporcionar insights accionables para reducir la tasa de Churn

## Limpieza y Tratamiento de Datos

El conjunto de datos original contenía información anidada en formato JSON que fue procesada y normalizada para su análisis:

1. Importación de datos: Se cargó el archivo JSON utilizando pandas.
2. Normalización de datos anidados: Las columnas anidadas ('customer', 'phone', 'internet', 'account') fueron expandidas para crear un DataFrame completo.
3. Limpieza de datos:
   - Estandarización de valores categóricos (convertidos a minúsculas y eliminación de espacios)
   - Reemplazo de valores inconsistentes en 'PaymentMethod'
   - Verificación de valores nulos (no se encontraron)
4. Exploración inicial: Se utilizaron métodos como `head()`, `info()` y `describe()` para entender la estructura y calidad de los datos.

El dataset final contiene 7267 entradas con 21 características por cliente, incluyendo información demográfica, servicios contratados, tipo de contrato, métodos de pago y cargos.

## Análisis Exploratorio de Datos

### Distribución del Churn

![Proporción de Churn](attachment:image.png)

La gráfica muestra que aproximadamente el 27% de los clientes abandonan el servicio (Churn = "yes"), mientras que el 73% permanece (Churn = "no"). Esta distribución indica que el problema de evasión es significativo y merece atención.

### Factores asociados al Churn

Se analizaron varias características para identificar su relación con el Churn:

1. Tipo de contrato:
   - Los clientes con contratos mensuales tienen una tasa de Churn mucho más alta (~43%) comparado con contratos anuales (~11%).

2. Servicio de internet:
   - Los clientes con fibra óptica muestran la mayor tasa de Churn (~42%), seguidos por DSL (~19%).

3. Método de pago:
   - Los pagos electrónicos (e_check) están asociados con mayor Churn (~34%).
   - Los pagos automáticos (bank_transfer, credit_card) tienen tasas más bajas (~16-17%).

4. Facturación sin papel:
   - Los clientes con PaperlessBilling tienen mayor probabilidad de abandonar (~34% vs ~16%).

5. Servicios adicionales:
   - Clientes sin servicios como seguridad online, backup online o protección de dispositivo tienen mayores tasas de Churn.

6. Dependientes:
   - Clientes sin dependientes tienen mayor probabilidad de abandonar (~31% vs ~16%).

7. Cargos mensuales:
   - Los clientes que abandonan tienden a tener cargos mensuales más altos

## Conclusiones e Insights

Los análisis revelan patrones claros entre los clientes que abandonan TelecomX:

1. Los clientes con contratos a corto plazo (mensuales) son más propensos a abandonar que aquellos con compromisos a más largo plazo.

2. La fibra óptica, aunque es un servicio más avanzado, presenta la mayor tasa de Churn, lo que podría indicar problemas con este servicio específico o expectativas no cumplidas.

3. Los métodos de pago automáticos están asociados con menor Churn, posiblemente por la comodidad y menor fricción.

4. La facturación sin papel está correlacionada con mayor Churn, lo que sugiere que estos clientes podrían estar menos comprometidos o tener preferencias diferentes.

5. La falta de servicios adicionales (seguridad, backup) aumenta el riesgo de Churn, indicando que estos servicios podrían aumentar el valor percibido y la retención.

6. Clientes sin dependientes son más propensos a abandonar, posiblemente porque tienen menos barreras para cambiar de proveedor.

## Recomendaciones Estratégicas

Basado en los hallazgos, se proponen las siguientes acciones para reducir el Churn:

1. Incentivar contratos a más largo plazo:
   - Ofrecer descuentos o beneficios exclusivos para contratos anuales.
   - Comunicar claramente las ventajas de los contratos a largo plazo.

2. Mejorar la experiencia con fibra óptica:
   - Investigar y resolver posibles problemas específicos de este servicio.
   - Asegurar que las expectativas de velocidad/calidad se cumplen.

3. Promover pagos automáticos:
   - Crear incentivos para que los clientes adopten pagos automáticos.
   - Reducir la fricción en el proceso de pago.

4. Revisar la estrategia de facturación sin papel:
   - Investigar por qué esta opción correlaciona con mayor Churn.
   - Potencialmente mejorar la experiencia de facturación digital.

5. Promover servicios adicionales:
   - Paquetes promocionales que incluyan servicios de seguridad/backup.
   - Comunicar el valor agregado de estos servicios.

6. Programas de fidelización para clientes sin dependientes:
   - Desarrollar ofertas específicas para este segmento de mayor riesgo.
   - Programas de recompensas por permanencia.

7. Sistema de alerta temprana:
   - Implementar modelos predictivos para identificar clientes en riesgo.
   - Intervenciones proactivas para estos clientes.

## Próximos Pasos

Para profundizar en el análisis, se podrían:
1. Desarrollar un modelo predictivo de Churn con machine learning.
2. Realizar análisis de cohortes para entender el Churn en diferentes periodos.
3. Segmentar los clientes con mayor riesgo para intervenciones personalizadas.
4. Realizar pruebas A/B con diferentes estrategias de retención.

```
